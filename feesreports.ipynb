```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(io="feecopy.xlsx",
                sheet_name="Student Master",
                usecols=[0,3,14,15,20])
overalltable=df.loc[df["Agrmnt No."].str.contains("N",na=False,case=False)]
ay=sns.countplot(data=df,y="Class",dodge=False)
ay.bar_label(ay.containers[0],fontsize=6)
ay.set_title("Total number of students")
plt.yticks(fontsize=6)
plt.xticks(fontsize=6)
plt.show()
```

```python
skey=pd.read_excel(io="feecopy.xlsx",
                sheet_name="Student Master",
                usecols=[3,15])
overalltable=skey.loc[skey["Agrmnt No."].str.contains("N",na=False,case=False)]
ax=sns.countplot(data=overalltable,y="Class")
ax.bar_label(ax.containers[0])
ax.set_title("Number of students without agreements")
plt.show()
```

```python
df["x"]=df["Agrmnt No."]
df.loc[(df.x!="N"),"x"]="Y"
az=sns.countplot(data=df,y="Class",hue="x")
az.bar_label(az.containers[0])
az.set_title("Merged Representation")
plt.show()
```

```python
dfn=df[df["Class"]=="N - AM"]
sns.countplot(data=dfn,x="Overall Due")

plt.show()
```
