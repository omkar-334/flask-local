```python
usecols=[0,1,2,3,5,9,13,17,21,25,29,33,37,38,39]
usenames=['Adm','Cell','Name','Class','Telugu', 'Hindi', 'English', 'Maths', 'Phy Sci', 'Bio Sci', 'Science','Social', 'Total', 'GPA', 'Grade']
useindex=["FA-1","FA-2","FA-3","FA-4","SA-1","SA-2"]
[sheet1,sheet2,sheet3,sheet4,sheet5,sheet6]=useindex

file1='markscopy1.xlsx'
table=pd.read_excel(
    io=file1,
    sheet_name=sheet1,
    usecols=[0,1,2,3],
    names=usenames,
    converters={'Cell':str,'Adm':str})
fa1=pd.read_excel(
    io=file1,
    sheet_name=sheet1,
    usecols=usecols,
    names=usenames,
    converters={'Cell':str,'Adm':str})
fa2=pd.read_excel(
    io=file1,
    sheet_name=sheet2,
    usecols=usecols,
    names=usenames,
    converters={'Cell':str,'Adm':str})
fa3=pd.read_excel(
    io=file1,
    sheet_name=sheet3,
    usecols=usecols,
    names=usenames,
    converters={'Cell':str,'Adm':str})
fa4=pd.read_excel(
    io=file1,
    sheet_name=sheet4,
    usecols=usecols,
    names=usenames,
    converters={'Cell':str,'Adm':str})
sa1=pd.read_excel(
    io=file1,
    sheet_name=sheet5,
    usecols=usecols,
    names=usenames,
    converters={'Cell':str,'Adm':str})
sa2=pd.read_excel(
    io=file1,
    sheet_name=sheet6,
    usecols=usecols,
    names=usenames,
    converters={'Cell':str,'Adm':str})

admvar="1005"


res=pd.concat([fa1,fa2,fa3,fa4,sa1,sa2])
res=res[res["Adm"]==admvar]
res=res.iloc[:,4:]
outlist=[Telugu, Hindi, English, Maths, PhySci, BioSci, Science,Social, Total, GPA, Grade]=[list(res['Telugu']),list(res['Hindi']),list(res['English']),list(res['Maths']),list(res['Phy Sci']),list(res['Bio Sci']),list(res['Science']),list(res['Social']),list(res['Total']),list(res['GPA']),list(res['Grade'])]
```
